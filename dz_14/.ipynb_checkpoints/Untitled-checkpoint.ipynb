1. Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
import pandas as pd
```

```python
rating = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
rating
```

```python
# (
#     rating
#     .groupby('userId').agg({'timestamp': ['max', 'min'], 'rating': 'count'})
#     .loc[x.rating['count'] > 100]
# )
```

```python
x = rating.groupby('userId').agg({'timestamp': ['max', 'min'], 'rating': 'count'})
x
```

```python
y = x.loc[x.rating['count'] > 100]
# y.set_index(['timestamp.max', 'timestamp.min', 'rating.count'])
y
```

```python
# y.loc[: , 'user_lifetime'] = y.timestamp['max'] - y.timestamp['min']
# y
```

```python
# def d_time(row):
#     return row.timestamp['max'] - row.timestamp['min']
```

```python
# y['user_lifetime'] = y.apply(d_time, axis=1)
# y
```

```python
y['user_lifetime'] = y.apply(lambda row: row.timestamp['max'] - row.timestamp['min'], axis=1)
y
```

```python
y.user_lifetime.mean()
```

- аналогичную таблицу по типам выручки с указанием адреса клиента

### К домашнему заданию, задача 2
Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
joined = rzd.merge(auto, how='outer').merge(air, how='outer')
joined.fillna(0)
```

```python
joinned_2 = client_base.merge(rzd, how='left').merge(auto, how='left').merge(air, how='left')
joinned_2.fillna(0)
```
