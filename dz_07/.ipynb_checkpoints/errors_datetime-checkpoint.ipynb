Дан поток дат в формате YYYY-MM-DD, в которых встречаются некорректные значения:
stream = [‘2018-04-02’, ‘2018-02-29’, ‘2018-19-02’]

Напишите функцию, которая проверяет эти даты на корректность. Т. е. для каждой даты возвращает True (дата корректна) или False (некорректная дата).

```python
from datetime import datetime
from datetime import timedelta
```

```python
def date_range(start_date, end_date):
    day_list = []
    d_1 = datetime.strptime(start_date, '%Y-%m-%d')
    d_2 = datetime.strptime(end_date, '%Y-%m-%d')
    while d_1 < (d_2 - timedelta(days=1)):
        d_1 += timedelta(days=1)
#         print(d_1)
#         day_list.append(d_1)
        day_list.append(d_1.strftime('%Y-%m-%d'))
#         day_list.append(str(d_1)[0:10])
    return day_list
```

```python
def date_range_2():
    try:
        day_list_2 = date_range(start_date, end_date)
    except ValueError:
        day_list_2 = []
    return day_list_2
```

Ваш коллега прислал код функции:

DEFAULT_USER_COUNT = 3

""“

element_to_delete = default_list[-1]

return default_list[DEFAULT_USER_COUNT-2]

При однократном вызове этой функции все работает корректно:
delete_and_return_last_user(1)
‘A101’

Однако, при повторном вызове получается ошибка IndexError: list index out of range.

Задание:

    Что значит ошибка list index out of range?
    Почему при первом запуске функция работает корректно, а при втором - нет?


```python
DEFAULT_USER_COUNT = 3

def delete_and_return_last_user(region, default_list=['A100', 'A101', 'A102']):
    """
    Удаляет из списка default_list последнего пользователя
    и возвращает ID нового последнего пользователя.
    """
    element_to_delete = default_list[-1]
    default_list.remove(element_to_delete)

    return default_list[DEFAULT_USER_COUNT-2]
```

```python
delete_and_return_last_user(1)
```

Функция в целом не корректна, так как после второго ее вызова в списке default_list остается один элемент, а возвратить она должна элемент с адресом 3-2=1, которого уже не существует и возникает ошибка индекс вне диапазона (в списке присутствует только 0-й индекс, а функция должна вернуть значение под 1-м индексом).

Немного преобразуем для наглядности:

```python
DEFAULT_USER_COUNT = 3
default_list=['A100', 'A101', 'A102']
def delete_and_return_last_user(default_list):
    element_to_delete = default_list[-1]
    default_list.remove(element_to_delete)

    return default_list[DEFAULT_USER_COUNT-2]
```
