# Проверим несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

```python
# сравним уровень удовлетворенности уволенных со средним уровнем удовлетворенности в генеральной совокупности
# используем одновыборочный t-тест Стьюдента
```

```python
stats.ttest_1samp(left['satisfaction_level'], data['satisfaction_level'].mean())
```

```python
# p value < 0.05, это значит что уровень удовлетворенности уволившихся отличается от всей совокупности
# возможно работники не удовлетворены работой по каким-то причинам
```

```python
# сравним последнюю оценку уволенных со средней последней оценкой в генеральной совокупности
# используем одновыборочный t-тест Стьюдента
```

```python
stats.ttest_1samp(left['last_evaluation'], data['last_evaluation'].mean())
```

```python
# p value > 0.05, это значит что последняя оценка уволившихся не отличается от всей совокупности
# возможно последняя оценка сотрудника не повлияла на его решение уволиться
```

```python
# сравним средние часы в месяц уволенных со средними часами в месяц в генеральной совокупности
# используем одновыборочный t тест Стьюдента
```

```python
stats.ttest_1samp(left['average_montly_hours'], data['average_montly_hours'].mean())
```

```python
# p value < 0.05, это значит что средние часы в месяц уволившихся отличаются от всей совокупности
# возможно уволившиеся сотрудники перерабатывали, либо наоборот работали мало
```

```python
# сравним стаж уволенных со средним стажем в генеральной совокупности
# используем одновыборочный t тест Стьюдента
```

```python
stats.ttest_1samp(left['time_spend_company'], data['time_spend_company'].mean())
```

```python
# p value < 0.05, это значит что стаж уволившихся отличается от среднего стажа всей совокупности
# возможно сотрудник, который как правило не задерживается в компании, покидает ее не отработав 
# среднестатистического стажа для данной компании
```

```python
# Сравним уровень удовлетворенности сотрудников по зарплатам для всей совокупности
# используем множественный тест ANOVA
```

```python
stats.f_oneway(data[data['salary']=='low']['satisfaction_level'], data[data['salary']=='medium']['satisfaction_level'], data[data['salary']=='high']['satisfaction_level'])
```

```python
# p value < 0.05, это значит что уровень удовлетворенности отличается по уровням зарплат
# сотрудники имеют разный уровень удовлетворенности в зависимости от уровня зарплаты
```

```python
# сравним уровень удовлетворенности уволенных с уровнем удовлетворенности работающих сотрудников
# используем двухвыборочный t критерий
```

```python
stats.ttest_ind(left['satisfaction_level'], working['satisfaction_level'])
```

```python
# p value < 0.05, это значит что уровень удовлетворенности уволенных отличается от уровня удовлетворенности 
# работающих
```

```python
# сравним стаж в компании уволенных со стажем работающих сотрудников
# используем двухвыборочный t критерий
```

```python
stats.ttest_ind(left['time_spend_company'], working['time_spend_company'])
```

```python
# p value < 0.05, это значит что стаж уволенных отличается от стажа работающих
```

```python
# сравним уровень удовлетворенности сотрудников со стажем больше среднего с уровнем удовлетворенности со стажем меньше среднего
# используем двухвыборочный t критерий
```

```python
stats.ttest_ind(data[data['time_spend_company']<data['time_spend_company'].mean()]['satisfaction_level'], data[data['time_spend_company']>data['time_spend_company'].mean()]['satisfaction_level'])
```

```python
# p value < 0.05, это значит что уровень удовлетворенности сотрудников со стажем больше среднего 
# отличается от уровня удовлетворенности сотрудников со стажем меньше среднего
```

```python
# Построим модель деревья решений
```

```python
output_y = data.left
input_x = data[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
# targets=(output_y.iloc[:,:]==1).idxmax(1)
targets=(output_y)
print(targets.value_counts())
Y=le.fit_transform(targets)
```

```python
from sklearn.model_selection import train_test_split
X_train_22, X_test_22, y_train_22, y_test_22 = train_test_split(input_x, Y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

# создаем модель деревья решений
# выбираем 100 деревьев в качестве параметра
model=RandomForestClassifier(n_estimators=100)

# обучаем модель
model.fit(X_train_22,y_train_22)
```

```python
headers = list(X_train_22.columns.values)

feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(6, 6))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Важность атрибутов')
plt.ylabel('Атрибуты')
plt.title("Наиболее важные атрибуты")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report
model_pred = model.predict(X_test_22)

print(classification_report(y_test_22, model_pred))

model.score(X_test_22, y_test_22)
```

```python
# Наиболее важным атрубутом, влияющим на увольнение сотрудников является уровень удовлетворенности,
# скорее всего этот показатель собирался на основе анкетирования сотрудников и является интегрирующим фактором.
# Также оказывает влияние номер проекта, возможно в разных проектах разная удовлетворенность сотрудников, 
# разная зарплата и нагрузка.
# карьерный рост и несчастные случаи не оказывали значительного влияния.
```
